K-Means is a clustering algorithm in machine learning that can group an unlabeled dataset very quickly and efficiently in just a few iterations. It works by labelling all instances on the cluster with the closest centroid. When the instances are centred around a particular point, that point is called a centroid.

K-Means Clustering using Python

The computational complexity of the K-Means clustering algorithm is generally linear concerning:

1. the number of instances m,

2. the number of clusters k,

3. and the number of dimensions n.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv")
print(data.columns)
```

```python
data = data.loc[:, ["median_income", "latitude", "longitude"]]
print(data.head())
```

```python
kmeans = KMeans(n_clusters=6)
data["Cluster"] = kmeans.fit_predict(data)
data["Cluster"] = data["Cluster"].astype("int")
print(data.head())
```

Now letâ€™s have a look at the clusters identified by the algorithm by using a scatterplot:



```python
plt.style.use("seaborn")
plt.rc("figure", autolayout=True)
plt.rc("axes", labelweight='bold', labelsize='large', titleweight='bold', titlesize=14, titlepad=10)
sns.relplot(x="longitude", y="latitude", hue="Cluster", data=data, height=6)
plt.show()
```

Conclusion:

The scatter plot above shows the geographic distribution of the clusters. 

It appears that the algorithm created separate segments for the high-income area.
